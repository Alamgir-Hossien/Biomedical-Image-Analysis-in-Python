# DICOM File Formate
DICOM stands for Digital Imaging and Communications in Medicine. DICOM files were introduced to maintain uniformity among varied types of medical image modalities. It is a standard format to view, store, share and retrieve medical images.

Biomedical images are generally found in DICOM image format (.dcm).

# pydicom module in python
Python offers a powerful module, pydicom to work with the DICOM files such as medical images, reports, and radiotherapy objects. pydicom reads modify and write data in DICOM files.Â 
Here a sample .dcm image from a Magnetic Resonance Imaging.

```python
'''This program read dicom file(D000.dcm) using pydicom module.
It will print patient information, meta data or other information contain D0001.dcm file'''

#Reading dicom file in python
#Displaying Patient information(Meta data)
import pydicom as dicom 
import matplotlib.pyplot as plt
x ='D0001.dcm'
read = dicom.dcmread(x)
print(read)
```

```python
'''In this program we will show the image contain in D0001.dcm file using pydicom and matplotlib module'''
import pydicom as dicom
import matplotlib.pyplot as plt
x ='D0001.dcm'
reading_image = dicom.dcmread(x)
plt.imshow(reading_image.pixel_array, cmap=plt.cm.gray)
```

# imagio module 
Imageio is a Python library that provides an easy interface to read and write a wide range of image data, including animated images, volumetric data, scientific formats, dicom format. It is cross-platform.

We can also work with dicom file in imageio module 

```python
'''In this program we read dicom file using imageio module in python'''
import imageio
import matplotlib.pyplot as plt
x ='D0008.dcm'
reading_dicom_file = imageio.imread(x)
print(reading_dicom_file.meta.keys())
print(reading_dicom_file.meta.values())
```

```python
'''In this program we will show images contain in D0008.dcm file using imageio module in python'''
import imageio
import matplotlib.pyplot as plt
x ='D0008.dcm'
reading_dicom_file = imageio.imread(x)
plt.imshow(reading_dicom_file)
```

```python
'''In this program we read multiple images(Folder image) in python'''
import imageio
from pathlib import Path
import matplotlib.pyplot as plt

images = list()
for file in Path("medical images").iterdir():
    if not file.is_file():
        continue

    images.append(imageio.imread(file))

fig, axes = plt.subplots(nrows=1, ncols=5)

print('Toltal images in folder: ',len(images))

for i in range(len(images)):
  axes[i].imshow(images[i], cmap='gray')


```

```python
'''In this program we will resize image'''
import imageio
import matplotlib.pyplot as plt
import cv2
path = 'mri.jpg'
read = cv2.imread(path)
r=cv2.resize(read, (400,200))
plt.imshow(r, cmap='gray')
```

```python
'''In this program we will rotate image'''
import cv2
import numpy as np
import matplotlib.pyplot as plt
x = 'mri3.jpg'
img = cv2.imread(x)
res = cv2.rotate(img, cv2.ROTATE_180)
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].imshow(img)
axes[1].imshow(res)
```

```python
'''This program will detect edges using Canny edge detection algorithm'''
import cv2
import numpy as np
import matplotlib.pyplot as plt
x = 'mri3.jpg'
img = cv2.imread(x)
edges = cv2.Canny(img,200,200)
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].imshow(img)
axes[1].imshow(edges, cmap='gray')
```

```python
'''This program will show the intensity of x-ray image'''
import imageio
import numpy as np
import matplotlib.pyplot as plt

# Load the hand radiograph
im = imageio.imread("hand_xray.jpg")
print('Data type:', im.dtype)
print('Min. value:', im.min())
print('Max value:', im.max())

# Plot the grayscale image
plt.axis('off')
plt.imshow(im, cmap='gray', vmin=0, vmax=255)
plt.colorbar()
```

```python
'''In this program will denoising MRI image using gaussian filter'''
from skimage import img_as_float
from matplotlib import pyplot as plt
from skimage import io
from scipy import ndimage as nd

noisy_img=img_as_float(io.imread("MRI_noisy.tif"))
ref_img=img_as_float(io.imread("MRI_noisy.tif"))
gaussian_img=nd.gaussian_filter(noisy_img,sigma=3)
plt.subplot(121)
plt.imshow(gaussian_img,cmap='gray')
plt.xlabel('Filtered image')
plt.subplot(122)
plt.imshow(noisy_img,cmap='gray')
plt.xlabel('Noisy image')
```
